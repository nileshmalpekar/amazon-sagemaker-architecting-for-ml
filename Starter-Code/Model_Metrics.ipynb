# Job 4000

```python
import boto3
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

client = boto3.client('logs')
lgn = '/aws/sagemaker/TrainingJobs'
lsn = 'Scientist-36-imageclassification-2019-02-14-15-31-12/algo-1-1550158464'
log = client.get_log_events(logGroupName=lgn,logStreamName=lsn)


trn_accs = []
val_accs = []


for e in log['events']:
    msg=e['message']

    if 'Validation-accuracy' in msg and 'INFO' not in msg:
        val = msg.split("=")
        val = val[1]
        val_accs.append(float(val))
    if 'Train-accuracy' in msg and 'INFO' not in msg:
        trn = msg.split("=")
        trn = trn[1]
        trn_accs.append(float(trn))
        

print("Maximum validation accuracy: %f " % max(val_accs))
plt.clf()
fig, ax = plt.subplots()
plt.xlabel('Epochs')
plt.ylabel('Accuracy')


#trn_plot, = ax.plot(range(epochs), trn_accs, label="Training accuracy")
#val_plot, = ax.plot(range(epochs), val_accs, label="Validation accuracy")

trn_plot, = ax.plot(trn_accs, label="Training accuracy")
val_plot, = ax.plot(val_accs, label="Validation accuracy")


plt.legend(handles=[trn_plot,val_plot])
ax.yaxis.set_ticks(np.arange(0.0, 1.05, 0.05))
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.2f'))
plt.show()
    
```
